```python
import json
import pandas as pd
import glob
```

```python
from re import sub
def normalize_snake_case(s):
  return '_'.join(
    sub('([A-Z][a-z]+)', r' \1',
    sub('([A-Z]+)', r' \1',
    s.replace('-', ' '))).split()).lower()
```

# Lendo os arquivos VRA

```python
import glob
list_vra_files =(glob.glob("VRA/*"))
list_DataFrame_vra = []
```

```python

for vra in list_vra_files:
    #namefile = "VRA/"+vra
    with open(vra, encoding='utf-8-sig') as fh:
       data = json.load(fh)
    
    list_DataFrame_vra.append(pd.DataFrame.from_dict(data))
    
vra_df = pd.concat(list_DataFrame_vra,ignore_index=True)    
```

```python
vra_df
```

```python
col_name_ant = vra_df.columns.values.tolist()
col_name_new = []

for i in col_name_ant:
    string= normalize_snake_case(i)
    col_name_new.append(string)
    
dictionary = dict(zip(col_name_ant, col_name_new))
vra_df =vra_df.rename(columns=dictionary)
```

```python
vra_df
```

```python
vra_df
vra_df = vra_df.dropna( axis=0, how='all')
vra_df.to_csv('vra.csv', index=False)
#Remoover os NAN
#Salvar o arquivo
```

 - Para cada aeroporto trazer a companhia aérea com maior atuação no ano com as seguintes informações:
    - Nome do Aeroporto
    - ICAO do Aeroporto
    - Razão social da Companhia Aérea
    - Quantidade de Rotas à partir daquele aeroporto
    - Quantidade de Rotas com destino àquele aeroporto
    - Quantidade total de pousos e decolagens naquele aeroporto

```python
q1 = """SELECT icao_aeródromo_destino, icao_empresa_aérea,strftime('%Y',partida_real) as Year, COUNT (*)  qnt_destino 
FROM vra_df
where Year = '2021' and situação_voo = 'REALIZADO'
group by icao_aeródromo_destino, icao_empresa_aérea 
order by qnt_destino DESC"""
```

```python
q2 = """SELECT icao_aeródromo_origem, icao_empresa_aérea,strftime('%Y',partida_real) as Year, COUNT (*)  qnt_origem
FROM vra_df
where Year = '2021' and situação_voo = 'REALIZADO'
group by icao_aeródromo_origem, icao_empresa_aérea 
order by icao_aeródromo_origem DESC"""
```

```python
df1= sqldf(q1, globals())
```

```python
df2=sqldf(q2, globals())
sqldf(q2, globals())
```

```python
q3 = """Select *  from df1 
join df2 on df2.icao_aeródromo_origem = df1.icao_aeródromo_destino and df2.icao_empresa_aérea = df1.icao_empresa_aérea

"""
```

```python
soma = sqldf(q3, globals())
```

```python
soma['qnt_rota'] = soma['qnt_destino']+soma['qnt_origem']
```

```python
q4 ="""SELECT 
    icao_aeródromo_destino as aero,
    icao_empresa_aérea,
    qnt_rota
FROM (
    SELECT 
        icao_aeródromo_destino, 
        icao_empresa_aérea,
        qnt_destino,
        qnt_origem,
        qnt_rota,
        ROW_NUMBER() OVER (
            PARTITION BY icao_aeródromo_destino 
            ORDER BY qnt_rota DESC
        ) RowNum
    FROM 
        soma )
WHERE
    RowNum = 1;"""
```

```python
qtd_rota =sqldf(q4, globals())
sqldf(q4, globals())
```

```python
qfinal = """Select * from qtd_rota 
left join soma on soma.icao_aeródromo_destino = qtd_rota.aero and soma.icao_empresa_aérea = qtd_rota.icao_empresa_aérea
"""
```

```python
qfinal = sqldf(qfinal, globals())
```
